```python
def country_data(country, status):
    response_country = requests.get('https://api.covid19api.com/total/country/{}/status/{}'.format(country, status))
    return response_country.json()
```

```python
def date_record_dict(country_data):
    return {datetime.strptime(d['Date'][:10], '%Y-%m-%d'): d['Cases'] for i,d in enumerate(country_data)}
```

```python
### This is for dash page
def country_status_dict(country, status):
    response_country = requests.get('https://api.covid19api.com/total/country/{}/status/{}'.format(country, status))
    country_data = response_country.json()
    case_country = {datetime.strptime(d['Date'][:10], '%Y-%m-%d'): d['Cases'] for i,d in enumerate(country_data)}
    return case_country
```

```python
def quick_confirmed_bar_plot(case_country, country_name):
    day_count = {i+1: case_country.get(c) for i, c in enumerate(case_country)}
    first_date = list(case_country.keys())[0].strftime("%Y-%m-%d")
    df_dict = pd.DataFrame.from_dict(day_count, orient='index', columns=['Cases'])
    df_dict.reset_index(inplace=True)
    df_dict.rename(columns={'index': 'Day'}, inplace=True)
    fig = px.bar(df_dict, x='Day', y='Cases', 
                  title = '# of Confirmed Cases in <b>' + country_name +'</b>')
    fig.update_xaxes(title_text='nth day ' + '(First recorded date is <b>{}</b>)'.format(first_date))
    fig.update_yaxes(title_text='# of Cases')
    return fig
```

```python
def avg_confirmed_bar_plot(case_country, country_name):
    day_count = {i+1: case_country.get(c) for i, c in enumerate(case_country)}
    first_date = list(case_country.keys())[0].strftime("%Y-%m-%d")
    df_dict = pd.DataFrame.from_dict(day_count, orient='index', columns=['Cases'])
    df_dict.reset_index(inplace=True)
    df_dict.rename(columns={'index': 'Day'}, inplace=True)
    df_dict['Infection rate (# of cases per 1M people)'] = df_dict['Cases']/country_pop.get(country_name)*1000000
    fig = px.bar(df_dict, x='Day', y='Infection rate (# of cases per 1M people)', 
                  title = '# of Confirmed Cases in <b>' + country_name +'</b>')
    fig.update_xaxes(title_text='nth day ' + '(First recorded date is <b>{}</b>)'.format(first_date))
    fig.update_yaxes(title_text='Infection rate (# of cases per 1M people)')
    return fig
```

### Feature Selection

- Filter Method: filtering our dataset and taking only a subset of it containing all the relevant features (eg. correlation matrix using Pearson Correlation).
- Wrapper Method: follows the same objective of the FIlter Method but uses a Machine Learning model as it’s evaluation criteria (eg. Forward/Backward/Bidirectional/Recursive Feature Elimination). p.s. this method can be more accurate than filtering but is more computationally expensive.
- Embedded Method: examines the different training iterations of our ML model and then ranks the importance of each feature based on how much each of the features contributed to the ML model training (eg. LASSO Regularization).

ref: https://towardsdatascience.com/feature-selection-techniques-1bfab5fe0784

```python
import pandas as pd
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
df = pd.read_csv('winequality-red.csv') # Load the data

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor

# The target variable is 'quality'.
Y = df['quality']
X =  df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 
         'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']]
# Split the data into train and test data:
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
# Build the model with the random forest regression algorithm:
model = RandomForestRegressor(max_depth=6, random_state=0, n_estimators=10)
model.fit(X_train, Y_train)
```

#### Variable Importance Plot 

```python
## variable importance plot
## The top variables contribute more to the model than the bottom ones and thus have high predictive power.

import shap
shap_values = shap.TreeExplainer(model).shap_values(X_train)
shap.summary_plot(shap_values, X_train, plot_type="bar",max_display=20, plot_size=(12,6))
#max_display is the number of top features showing in the graph
```

```python
def ABS_SHAP(df_shap,df):
    #import matplotlib as plt
    # Make a copy of the input data
    shap_v = pd.DataFrame(df_shap)
    feature_list = df.columns
    shap_v.columns = feature_list
    df_v = df.copy().reset_index().drop('index',axis=1)
    
    # Determine the correlation in order to plot with different colors
    corr_list = list()
    for i in feature_list:
        b = np.corrcoef(shap_v[i],df_v[i])[1][0]
        corr_list.append(b)
    corr_df = pd.concat([pd.Series(feature_list),pd.Series(corr_list)],axis=1).fillna(0)
    # Make a data frame. Column 1 is the feature, and Column 2 is the correlation coefficient
    corr_df.columns  = ['Variable','Corr']
    corr_df['Sign'] = np.where(corr_df['Corr']>0,'red','blue')
    
    # Plot it
    shap_abs = np.abs(shap_v)
    k=pd.DataFrame(shap_abs.mean()).reset_index()
    k.columns = ['Variable','SHAP_abs']
    k2 = k.merge(corr_df,left_on = 'Variable',right_on='Variable',how='inner')
    k2 = k2.sort_values(by='SHAP_abs',ascending = True)
    colorlist = k2['Sign']
    ax = k2.plot.barh(x='Variable',y='SHAP_abs',color = colorlist, figsize=(12,6),legend=False)
    ax.set_xlabel("SHAP Value (Red = Positive Impact)")
    
ABS_SHAP(shap_values,X_train) 
```

```python
## The SHAP value plot
## show the positive and negative relationships of the predictors with the target variable
## Color Explanation: the color (Feature value) means the feature value of each record (Red is relatively high and vice versa)
## so we could tell 'alcohol' has more clear correlation between feature value and SHAP value but 'density' is not
## Horizontal Explanation: The location shows the whether the effect of that value is associated with a higher/lower prediction.

### Insight: “volatile acidity” is negatively correlated with the target variable
shap.summary_plot(shap_values, X_train, max_display=20, plot_size=(14,6))
```

#### SHAP Dependence Plot
The partial dependence plot shows the marginal effect one or two features have on the predicted outcome of a machine learning model (J. H. Friedman 2001). It tells whether the relationship between the target and a feature is linear, monotonic or more complex. (Most important, it could see if the effect is heavily affected by other features)

##### About the color (interaction with other features)
If not assigning interaction_index, it is used to pick what seems to be the strongest interaction (SHAP interaction values)

In this case, sulphates interact with alcohol the most (This means that the value of alcohol would affect contribution of the sulphates the most)
p.s. When alcohol is low, the lower sulphates could reduce the quality (not causation, could be other reason!) However, when alcohol is high, the higher sulphates increase the quality (though the effect is not as clear as when lower alcohol)

https://christophm.github.io/interpretable-ml-book/shap.html#shap-dependence-plot
##### Better than other PDP
-  Traditional PDPs assume that the target features are independent from the complement features, and this assumption is often violated in practice.



```python
# partial dependence plots
import matplotlib.pyplot as pl
import matplotlib

figsize = (12, 8) ## originally (7.5, 5)
fig = pl.figure(figsize=figsize)
ax = fig.gca()
shap.dependence_plot('alcohol', shap_values, X_train, interaction_index = 'auto'
                     ,dot_size=20, x_jitter=0, alpha=0.9, title=None, ax=fig.gca())

# possible title: 'SHAP Dependence Plot with Target & Feature_name'
```
